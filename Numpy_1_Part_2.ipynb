### Numpy arrays are faster

it takes 1.7 micro seconds to create a python array of squares:

```python
a = [1, 2, 3, 4, 5]
%timeit [i**2 for i in a]
```

But it takes 622 nano seconds for the same in numpy:

```python
import numpy as np
n = np.array(a)
%timeit n ** 2 #Broadcasting operation
```
