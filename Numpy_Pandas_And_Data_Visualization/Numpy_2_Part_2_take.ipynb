### Array in n-th dimension

given array: A = [*a<sub>ijk..<sub>*]

how do you iterate through n<sup>th</sup> dimension of an array ?

```python
import numpy as np

def letter_vector_rec(size:int, vector:np.ndarray):
    if not len(vector):
        return np.array([f'{i}' for i in range(size)])
    else:
        return np.array([f'{i}'+vector for i in range(size)])

def letter_vector(prefix:str, shape:list[int]):
    result = np.array([])
    for size in shape[::-1]:
        result = letter_vector_rec(size, result)
    return prefix + result

def extract_axis(arr, axis):
    for i in range(arr.shape[axis]):
        slice_i = np.take(arr, i, axis=axis)
        print(f'axis={axis}, index={i}:')
        print(slice_i, '\n')
```

```python
# for 3 dimensions
arr = letter_vector('a_', shape=(2,3,4))
arr
```

```python
extract_axis(arr, 0)
```

```python
extract_axis(arr, 1)
```

```python
extract_axis(arr, 2)
```
